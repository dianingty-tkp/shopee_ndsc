```python
# Start with loading all necessary libraries
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import re

import matplotlib.pyplot as plt
% matplotlib inline
```

```python
def prep(product_name) :
    clean = product_name.apply(lambda x : re.sub("[^A-Za-z0-9]+", ' ', str(x)).lower())
    product_name = list(clean)
    titles = ""
    for i in product_name:
        titles += " " + str(i)
    return titles
```

```python
# Load in the dataframe
train_data = pd.read_csv("../../data/new_training_set.csv", index_col=0)
test_data = pd.read_csv("../../data/new_test_sample.csv", index_col=0)
```

```python
train_data.head()
```

```python
# train data
train_title = prep(train_data["title_1"]) + " " + prep(train_data["title_2"])

wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(train_title)
plt.figure(figsize=[15, 20])
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
# train data
test_title = prep(test_data["title_1"]) + prep(test_data["title_2"])

wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(test_title)
plt.figure(figsize=[15, 20])
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
